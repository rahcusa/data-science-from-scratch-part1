```python
import matplotlib.pyplot as plt
```

```python
# will print matplotlib plots inline, only works in jupyter notebooks
%matplotlib inline
```

```python
# lets make some data to work with
import numpy as np
x = np.arange(11)
y = x ** 2
print(x)
print(y)
```

**1) Basic plotting**
We have some data in x and y, lets create a basic plot

```python

plt.plot(x, y)
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Figure/plot Title')
```

Creating multiple plots on the same canvas - use subplot()!

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1, 3, 1)
plt.plot(x, y, 'r--')

plt.subplot(1, 3, 2)
plt.plot(y, x, 'g*-')

plt.subplot(1, 3, 3)
plt.plot(x, y, 'r--')
plt.plot(y, x, 'g*-')

# lets adjust the layout
plt.tight_layout()
```

**2) Using Object-oriented approach for plotting**
Lets create a figure instance and add axes to that figure

```python
fig1 = plt.figure()

# step 2 - add axes
axes = fig1.add_axes([0.1, 0.1, 0.9, 0.9]) # add_axes([left, bottom, width, height])

# step 3 - plot the data on the axes we have created
axes.plot(x, y, 'b')

# We can add labels and titles as well
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Figure title')
```

We can add more axes to the figure (i.e. create an inset plot)

```python
fig2 = plt.figure()

axes_main = fig2.add_axes([0.1, 0.1, 0.9, 0.9])
axes_inset = fig2.add_axes([0.3, 0.5, 0.3, 0.4])

axes_main.plot(x, y, 'b')
axes_main.set_xlabel('X Label main')
axes_main.set_ylabel('Y Label main')
axes_main.set_title('Figure title main')

axes_inset.plot(y, x, 'r')
axes_inset.set_xlabel('X Label inset')
axes_inset.set_ylabel('Y Label inset')
axes_inset.set_title('Title inset')
```

Creating a figure and subplots - using OOP

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
plt.tight_layout()
```

```python
# we can share the x and y axes
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True)
plt.tight_layout()
```

```python
fig1, axes1 = plt.subplots(2, 1)
fig2, axes2 = plt.subplots(1, 2)
```

```python
type(axes1), len(axes1)
```

```python
# axes1 is a numpy array and we can iterate through them using a for loop
for ax in axes1:
    ax.plot(x, y, 'b')
    ax.set_xlabel('X label')
    ax.set_ylabel('Y label')
    ax.set_title('Title')
fig1.tight_layout()
fig1
```

Generating boxplots

```python
# generate data using a list comprehension
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data, vert=True, patch_artist=True)
```
